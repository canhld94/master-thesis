```python
database_dir = '/home/canhld/workplace/Daejeon_street_view/datasets/daejeon_sv_v2/'
import os
import shutil 
with open(os.path.join(download_dir, 'index.csv')) as f:
    contents = f.readlines()
lines = [x.strip('\n') for x in contents]
for line in lines:
    name = line.split(',')[1].strip(' ')
    for f in os.listdir(os.path.join(download_dir,name)):
        _, extension = f.split('.')
        if extension.lower() == 'jpg':
            new_file_name = name + '_' + f
            print(new_file_name)
            shutil.copyfile(os.path.join(download_dir,name,f), os.path.join(database_dir, new_file_name))
```
